```python
import matplotlib.pyplot as plt


# Create a title
plt.title("SPAM EMAIL PREDICTION", color="grey", size=40)

# Create a title and axis labels
mylabels = ['Not Spam', 'Spam']

#wedge spam from piechart
myexplode = [0.1, 0]

#change color of pie
mycolors = ['m', 'y']




email_count.plot(kind="pie", labels=mylabels, startangle = 180, autopct="%1.0f%%", explode = myexplode, shadow = True, colors = mycolors, figsize=(10, 8), textprops={"color": "w"})
plt.show()
```

It is concluded that 29% of the email is regarded as spam.
